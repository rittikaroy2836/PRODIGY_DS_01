```python
# ðŸ“Œ Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Dataset URL
url = "https://raw.githubusercontent.com/Prodigy-InfoTech/data-science-datasets/main/Task%201/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv"

# Skip the first 4 rows (metadata in World Bank datasets)
data = pd.read_csv(url, skiprows=4)

# Preview
print(data.head())
print(data.info())
# Extract country names and 2020 population
pop_2020 = data[['Country Name', '2020']].dropna()
pop_2020.head()
plt.figure(figsize=(8,6))
plt.hist(pop_2020['2020'], bins=30, color='skyblue', edgecolor='black')
plt.title("Population Distribution Across Countries (2020)")
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.show()
# Top 10 countries
top10 = pop_2020.sort_values('2020', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='2020', y='Country Name', data=top10, palette='viridis')
plt.title("Top 10 Most Populated Countries in 2020")
plt.xlabel("Population")
plt.ylabel("Country")
plt.show()
```
